# Down sampling

https://docs.obspy.org/packages/autogen/obspy.core.stream.Stream.decimate.html#obspy.core.stream.Stream.decimate

```python
from obspy import read

st = read()
st.plot()
print(st)
```

```python
st.decimate(2, no_filter=False, strict_length=False)
st.plot()
print(st)
```

# Over sampling (Interpolation)

https://docs.obspy.org/packages/autogen/obspy.core.trace.Trace.interpolate.html#obspy.core.trace.Trace.interpolate

```python
from obspy import read

st = read()
st.plot()
print(st)
```

```python
st.interpolate(sampling_rate=200,
               method='weighted_average_slopes',
               starttime=None,
               npts=None,
               time_shift=0.0)
st.plot()
print(st)
```

# Resample

https://docs.obspy.org/packages/autogen/obspy.core.trace.Trace.resample.html#obspy.core.trace.Trace.resample

```python
from obspy import read

st = read()
st.plot()
print(st)

st.resample(sampling_rate=70, window='hanning', no_filter=True, strict_length=False)
```
