```python
from obspy.signal.filter import envelope
import numpy as np
import matplotlib.pyplot as plt

import obspy
import obspy.signal


st = obspy.read("https://examples.obspy.org/RJOB_061005_072159.ehz.new")
print(st)
tr = st[0]
```

```python
# Envelope of data
data_envelope = envelope(tr.data)

# The plotting, plain matplotlib
a = 82
t = tr.times()
plt.plot(t, tr.data, 'k')
plt.plot(t, data_envelope, 'k:')
plt.title(tr.stats.starttime)
plt.xlim(a, a+3)
plt.show()
```

```python
# Envelope of filtered data
tr_filt = tr.copy()
tr_filt.filter('bandpass', freqmin=0.5, freqmax=8)
data_envelope_filt = envelope(tr_filt.data)

# The plotting, plain matplotlib
a = 82
t = tr_filt.times()
plt.plot(t, tr_filt.data, 'k')
plt.plot(t, data_envelope_filt, 'k:')
plt.title(tr_filt.stats.starttime)
plt.xlim(a, a+3)
plt.show()
```
