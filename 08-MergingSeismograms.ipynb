```python
from obspy import read
st = read('./sample-data/merge.msd')
st.plot()
print(st)
```

# Some change on data

```python
st0 = st.copy()

for tr in st0:
    tr.data += 10
st0[0].data[1000] = 8
st0.plot()
print(st0)
```

# Bad operation

```python
st01 = st0.copy()
st01.detrend('constant')
st01.merge()
st01.plot()
print(st01)
```

# Better operation

```python
st02 = st0.copy()
st02.merge(method=-1)
st02.detrend('constant')
st02.merge()
st02.plot()
print(st02)
```
