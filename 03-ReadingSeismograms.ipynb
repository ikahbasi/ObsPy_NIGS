# Slice data

```python
from obspy import read
from obspy import UTCDateTime as utc

st = read('./sample-data/detrend.msd')
st.plot()
print('Origin data:\n', st)

stime = utc('2021-05-17T00:54:00')
etime = utc('2021-05-17T00:55:00')
new_st = st.slice(starttime=stime, endtime=etime)
new_st.plot()
print('After cut:\n', new_st)
```
