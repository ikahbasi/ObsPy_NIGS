# Make obspy Trace

```python
from obspy import UTCDateTime, read, Trace, Stream
import numpy as np

# Convert to NumPy character array
data1 = np.array([1, 2, -1, 5, -10, 2])

# Fill header attributes
stats1 = {'network': 'BW', 'station': 'RJOB', 'location': '',
         'channel': 'WLZ', 'npts': len(data1), 'sampling_rate': 0.1,
         'mseed': {'dataquality': 'D'}}

# set current time
stats1['starttime'] = UTCDateTime()
tr1 = Trace(data=data1, header=stats1)
tr1.plot()
print(tr1)
```

# Make another obspy Trace

```python
from obspy import UTCDateTime, read, Trace, Stream

# Convert to NumPy character array
data2 = np.array([1, -5, 10, 0, 5, -10, 2])

tr2 = Trace(data=data2)
tr2.stats.network = 'BW'
tr2.stats.station = 'RJOB'
tr2.stats.channel = 'WLE'
tr2.stats.npts = len(data2)
tr2.stats.sampling_rate = 0.1
tr2.stats.mseed = {'dataquality': 'D'}
tr2.stats.starttime = UTCDateTime()
# set current time
tr2.plot()
print(tr2)
```

# Make obspy Stream

```python
st = Stream([tr1, tr2])
st.plot()
print(st)
```
