```python
from obspy import read

st = read()
tr = st[0]

tr.spectrogram()  
print(tr)
```

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
# import matplotlib.pyplot as plt
import numpy as np
cmap = plt.get_cmap(name='jet')


tr.spectrogram(per_lap=0.9,
               wlen=None,
               log=False,
               outfile=None,
               fmt=None,
               axes=None,
               dbscale=False,
               mult=8,
               cmap=cmap,
               zorder=None,
               title=None,
               show=True,
               sphinx=True,
               clip=[0.0, 1.0])

gradient = np.linspace(0, 1, 100)
plt.imshow(gradient.reshape(-1, 2), aspect='auto', cmap=plt.get_cmap('jet'))
```
