# Type of variables

You can use type() function to find type of variables.

type(variable)

## numbers

```python
# int
a1 = 1
a2 = 1.0
a3 = 1j

print('a1:', a1, type(a1))
print('a2:', a2, type(a2))
print('a3:', a3, type(a3))
```

# String

```python
b1 = 'ngis.ir'

print('b1:', b1, type(b1))
```

# NumPy

```python
import numpy as np
f = 1
t = np.linspace(0, 10, 1000)
a = np.sin(2*np.pi*f*t)
```

# Matplotlib

```python
import matplotlib.pyplot as plt
plt.plot(t, a)
plt.xlabel('Amplitude')
plt.ylabel('Time')
plt.show()
```

```python
plt.hist(a)
plt.show()
```
