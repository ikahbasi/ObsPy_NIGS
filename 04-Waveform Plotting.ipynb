```python
from obspy import read
st = read('sample-data/detrend.msd')
st.plot()
print(st)
```

```python
from obspy import Stream
tr = st[0].copy()

tr.detrend()
tr.trim(pad=True, endtime=tr.stats.endtime+1000, fill_value=0)
st2 = Stream(traces=[tr.copy() for ii in range(5)])
for ind, tr in enumerate(st2):
    tr.stats.distance = ind
    tr.stats.station = str(ind)
st2.plot()
st2.plot(type='section')
```
