```python
st = read('./sample-data/ppsd2.mseed')
inv = read_inventory('./sample-data/ppsd.xml')
```

```python
tr = st.select(id="BW.KW1..EHZ")[0]
```

```python
ppsd = PPSD(tr.stats, metadata=inv)
ppsd.add(st)
```

```python
ppsd.plot()
print(ppsd.times_processed)
```

```python
from obspy.imaging.cm import pqlx
ppsd.plot(cmap=pqlx)
```

```python
ppsd.plot_temporal([0.1, 3, 10])
```
