```python
st = read('./sample-data/ppsd2.mseed')
inv = read_inventory('./sample-data/ppsd.xml')
```

```python
tr = st.select(id="BW.KW1..EHZ")[0]
```

```python
ppsd = PPSD(tr.stats, metadata=inv)
ppsd.add(st)
```

```python
ppsd.plot()
print(ppsd.times_processed)
```

```python
from obspy.imaging.cm import pqlx
ppsd.plot(cmap=pqlx)
```

```python
ppsd.plot_temporal([0.1, 3, 10])
```

```python
ppsd.plot_spectrogram()
```

```python
from obspy import read
from obspy.signal import PPSD
st = read("https://examples.obspy.org/BW.KW1..EHZ.D.2011.038")
print(st)

paz = {'gain': 60077000.0,

       'poles': [-0.037004+0.037016j, -0.037004-0.037016j,

                 -251.33+0j, -131.04-467.29j, -131.04+467.29j],

       'sensitivity': 2516778400.0,

       'zeros': [0j, 0j]}
```

```python
tr = st[0]
ppsd = PPSD(tr.stats, paz)
ppsd.add(st)
print(ppsd.id)
ppsd.plot() 
```

```python
ppsd.save_npz(filename='./sample-data/ppsd.npz')
```

```python
from obspy.imaging.cm import pqlx

ppsd2 = PPSD.load_npz(filename='./sample-data/ppsd.npz')
ppsd2.plot(cmap=pqlx, show_mode=True, show_mean=True)
```

```python
ppsd.plot_spectrogram()
```
