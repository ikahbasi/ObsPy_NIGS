```python
from obspy.core import read

st = read("https://examples.obspy.org/ev0_6.a01.gse2")

st = st.select(component="Z")

tr = st[0]
```

```python
from obspy.signal.trigger import classic_sta_lta
from obspy.signal.trigger import plot_trigger
cft = classic_sta_lta(tr.data, int(5 * tr.stats.sampling_rate), int(10 *  tr.stats.sampling_rate))

plot_trigger(tr, cft, 1.5, 0.5)
```

Z Detect

```python
from obspy.signal.trigger import z_detect

cft = z_detect(tr.data, int(10 * tr.stats.sampling_rate))

plot_trigger(tr, cft, -0.4, -0.3)
```

```python
from obspy.signal.trigger import recursive_sta_lta

cft = recursive_sta_lta(tr.data, int(5 * tr.stats.sampling_rate), int(10 * tr.stats.sampling_rate))

plot_trigger(tr, cft, 1.2, 0.5)

```

```python
from obspy.signal.trigger import carl_sta_trig

cft = carl_sta_trig(tr.data, int(5 * tr.stats.sampling_rate), int(10 * tr.stats.sampling_rate), 0.8, 0.8)

plot_trigger(tr, cft, 20.0, -20.0)
```
